This dataset describes various attributes of residential properties involved in sales, with a focus on their structural characteristics, location, and sale conditions. Here's a summary of the key features:

### Dwelling Characteristics:
- **MSSubClass**: Classifies the type of dwelling (e.g., 1-story, 2-story, duplex, PUDs) and their construction year.
- **BldgType**: Differentiates dwelling types such as single-family, duplex, and townhouses.
- **HouseStyle**: Describes architectural styles, including stories and finishes.

### Structural Features:
- **OverallQual & OverallCond**: Rates the quality and condition of the house on a scale of 1 to 10.
- **YearBuilt & YearRemodAdd**: Indicates construction and remodeling years.
- **RoofStyle & RoofMatl**: Details the roof type and materials used.
- **Exterior1st & Exterior2nd**: Specifies exterior covering materials.
- **Foundation**: Types of foundation used.
- **Heating & HeatingQC**: Describes the heating type and its quality.

### Interior Features:
- **TotalBsmtSF**: Total basement area.
- **GrLivArea**: Above-grade living area in square feet.
- **KitchenQual**: Quality rating of the kitchen.

### Lot and Zoning Information:
- **LotFrontage & LotArea**: Dimensions of the lot.
- **MSZoning**: Zoning classification (e.g., residential, commercial).
- **LotShape & LandContour**: Describes lot shape and property flatness.

### Utilities and Accessibility:
- **Utilities**: Type of utilities available (e.g., all public, no sewer).
- **Street & Alley**: Type of road access and alley access.
- **CentralAir**: Indicates the presence of central air conditioning.

### Location and Proximity:
- **Neighborhood**: Identifies the neighborhood within city limits.
- **Condition1 & Condition2**: Proximity to various conditions like railroads or arterial streets.

### Sale Information:
- **MoSold & YrSold**: Month and year the property was sold.
- **SaleType**: Type of sale (e.g., conventional, cash).
- **SaleCondition**: Conditions surrounding the sale (e.g., normal, abnormal).

### Miscellaneous Features:
- **Fireplaces**: Number of fireplaces and their quality.
- **GarageType & GarageCars**: Details about the garage and its capacity.
- **MiscFeature**: Any miscellaneous features not categorized elsewhere.

This dataset provides comprehensive details that can be used for property evaluation, market analysis, and real estate modeling.

# 9. Treatment of Missing Values

## a. Replace Numeric Missing Values with Mean

```python
mean_imputed_data = df.copy()
for col in numeric_cols:
    mean_value = mean_imputed_data[col].mean()
    mean_imputed_data[col].fillna(mean_value, inplace=True)
```

```python
mean_imputed_data[numeric_cols].head()
```

## b. Replace Numeric Missing Values with Median

```python
median_imputed_data = df.copy()
for col in numeric_cols:
    median_value = median_imputed_data[col].median()
    median_imputed_data[col].fillna(median_value, inplace=True)
```

```python
median_imputed_data[numeric_cols].head()
```

## c. Drop Numeric Columns with Missing Values

```python
dropped_data = df.dropna(axis=1, thresh=df.shape[0] * 0.6)
```
